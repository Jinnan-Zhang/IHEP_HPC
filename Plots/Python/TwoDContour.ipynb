```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('Style/Paper.mplstyle')
from myploter import  show_2d_contour_and_projection


Npts=50
x=np.linspace(-15,15,Npts)
y=np.linspace(-15,15,Npts)

xx,yy=np.meshgrid(x,y)

x_data=np.reshape(xx,(Npts*Npts))
y_data=np.reshape(yy,(Npts*Npts))
z_data=x_data**2+y_data**2

nominal_x=x_data[[np.where(x_data>=0)[0][0]]]
nominal_y=y_data[[np.where(y_data>=0)[0][0]]]

show_2d_contour_and_projection(data_x=x_data,data_y=y_data,data_z=z_data,nominal_xs=nominal_x,nominal_ys=nominal_y, levels=[20,80,150],cmap='',save_file_name='')
```
